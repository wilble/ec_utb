```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

# we can either use an object oriented approach or a pyplot oriented approach. The recommended approach is the object oriented one whereas the pyplot approach is for quick interactive work if you are familiar with it

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

# The figure is the complete frame containing all axes, axis, artist. Axes gives us the region of our plotting and are named. Axis gives us the scale and limits of our plotting. Artist is everything visible on the figure and are generally not transferrable between axes 

### When plotting in Matplotlib, what is the expected input data type?

# The expected input data type in Matplotlib is numpy.array and the data type should be converted to numpy.array if possible

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python

x = np.linspace(-4,4,100) 
y = x**2 

fig, ax1 = plt.subplots()
ax1.set_title("Min första graf")
ax1.plot(x, y)


```

```python

 
x = np.linspace(-4,4,100) 
y = x**2 
   
plt.plot(x, y) 
  
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('Min andra graf') 
  
plt.show() 
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, (ax1, ax2) = plt.subplots(1,2, layout="constrained")
fig.suptitle("riktigt feta grafer")
ax1.scatter(random_data_x, random_data_y)
ax2.bar(names, values)
```
